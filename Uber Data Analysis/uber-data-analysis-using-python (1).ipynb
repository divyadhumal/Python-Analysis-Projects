```python
df.drop(columns=['Incomplete Rides','Incomplete Rides Reason','Cancelled Rides by Customer','Reason for cancelling by Customer','Cancelled Rides by Driver','Driver Cancellation Reason'])
```

```python
df['DateTime']= pd.to_datetime(df['Date'].astype(str) +' ' + df['Time'].astype(str), errors='coerce')
 
```

```python
print(df['DateTime'].dtype)
 
```

```python
df['hour']=df['Date'].dt.hour
df['day']=df['Date'].dt.day
df['weekday']=df['Date'].dt.weekday
df['month']=df['Date'].dt.month
```

```python
print(df[['Date','Time','DateTime','hour','weekday']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['hour'].value_counts().sort_index().plot(kind='bar', figsize=(10,5))
plt.title('Rides per Hour')
plt.show()

df['weekday'].value_counts().sort_index().plot(kind='bar')
plt.title('Rides per weekday(0=Mon)')
plt.xlabel('Weekday (0-Mon)')
plt.ylabel('Number of Rides')
plt.show()

df['month'].value_counts().sort_index().plot(kind='bar')
plt.title('Rides per Month')
plt.xlabel('Month(1=Jan)')
plt.ylabel('Number of Rides')
plt.show()
```

```python
rides_by_day_hour=df.groupby(['weekday','hour']).size().unstack()
sns.heatmap(rides_by_day_hour, cmap='YlGnBu')
plt.title("Heatmap of Rides by Day & Hour")
plt.show()
```
